2.  What is an interpreter in Python ?

    Ans. An interpreter in Python is a program that reads and executes Python code line by line instead of converting the whole program into machine code at once.

    Simple Definition

    A Python interpreter is software that translates Python code into instructions the computer can understand and runs them immediately.

    How it Works

    You write code in Python (a .py file).

    The Python interpreter reads the code one line at a time.

    It converts the code into bytecode.

    The Python Virtual Machine (PVM) executes that bytecode.

    Example

    Python code:

    print("Hello World")

    Interpreter process:

    Reads print("Hello World")

    Translates it

    Executes it immediately

    Output: Hello World

    Key Features of the Python Interpreter

    Executes code line by line

    Helps debug easily (errors apRpear immediately)

    Supports interactive mode (you can run commands directly)

    Example interactive mode:

    >>> 5 + 3

    8

    Example of Python Interpreter

    The most common interpreter is CPython, which is the official implementation of Python.

6. Why are lists mutable, but tuples are immutable

    Ans.In Python, lists are mutable and tuples are immutable by design to serve different purposes in programming.

    Lists are Mutable

    A list can be changed after it is created. You can add, remove, or modify elements.

    Example:

    my_list = [1, 2, 3]
    
    my_list[0] = 10

    print(my_list)

    Output:

    [10, 2, 3]

    This makes lists useful when you need flexible data that may change during program execution.

    Tuples are Immutable

    A tuple cannot be changed after it is created.

    Example:

    my_tuple = (1, 2, 3)
    
    my_tuple[0] = 10   # Error

    This will produce an error because tuples do not allow modification.

    Why Python Designed Them This Way

    Data Safety – Tuples protect data from accidental changes.

    Performance – Tuples are generally faster and use less memory than lists.

    Hashability – Tuples can be used as dictionary keys, but lists cannot.

7. What is the difference between “==” and “is” operators in Python ?

    Ans.In Python, the == and is operators are used for comparison, but they work differently.

    I. == (Equality Operator)

    Checks whether the values of two objects are equal.

    Example:

    a = [1, 2, 3]
    
    b = [1, 2, 3]

    print(a == b)

    Output:

    True

    Both lists have the same values, so == returns True.

    II. is (Identity Operator)

    Checks whether two variables refer to the same object in memory.

    Example:

    a = [1, 2, 3]
    
    b = [1, 2, 3]

    print(a is b)

    Output:

    False

    Even though values are the same, they are stored in different memory locations.

10. What is the difference between implicit and explicit type casting ?

    I. Implicit Type Casting

    Automatically done by Python.

    No programmer intervention is required.

    Example:

    a = 5      # int
    
    b = 2.5    # float
    
    c = a + b
    
    print(c)

    Output: 7.5

    II. Explicit Type Casting

    Manually done by the programmer using functions like int(), float(), str().

    Example:

    x = "10"
    
    y = int(x)
    
    print(y)

    Here, the string is manually converted into an integer.

12.  How does the elif statement work ?

    Ans. In Python, the elif (else if) statement is used to check multiple conditions after an if statement.

    How it Works

    Python first checks the if condition.

    If it is False, Python checks the elif condition.

    If the elif condition is True, its block of code runs.

    If none of the conditions are true, the else block runs.

    Example:

    x = 10

    if x > 20:
    
    print("Greater than 20")

    elif x > 5:
    
    print("Greater than 5")
    
    else:
    
    print("5 or less")

    Output:

    Greater than 5

13. What is the difference between for and while loops ?

    Ans.In Python, both for and while loops are used to repeat a block of code, but they work differently.

    I. for Loop

    Used when the number of iterations is known.

    Iterates over a sequence like a list, tuple, or range.

    Example:

    for i in range(5):
    
    print(i)

    II. while Loop

    Used when the number of iterations is not known.

    Runs as long as the condition is True.

    Example:

    i = 0
    
    while i < 5:
    print(i)
    i += 1

    def modify_values(num, my_list):
